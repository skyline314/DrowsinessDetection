```python
import os

folder_path = "YawDD dataset/yawn" 

start_num_old = 452
end_num_old = 726

start_num_new = 1

print(f"Memproses file di folder: {folder_path}")
print("=============================================")

count = 1
for i in range(start_num_old, end_num_old + 1):
    
    old_filename = f"{i}.jpg"
    old_path = os.path.join(folder_path, old_filename)
    
    if os.path.exists(old_path):
        new_frame_id = count
        count += 1
        
        new_filename = f"{subject_id}-{new_frame_id}.jpg"
        new_path = os.path.join(folder_path, new_filename)
        os.rename(old_path, new_path)
        
        print(f"Berhasil: Mengubah '{old_filename}' -> '{new_filename}'")
    else:
        print(f"Peringatan: File '{old_filename}' tidak ditemukan.")

print("\n=============================================")
print("Proses pengubahan nama file selesai.")

subject_id += 1
```
